```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
sr = 16000
duration = 3  # seconds
t = np.linspace(0, duration, int(sr * duration), endpoint=False)

# Audio Sample 1 (speech-like)
audio1 = (
    0.5 * np.sin(2 * np.pi * 200 * t) +
    0.3 * np.sin(2 * np.pi * 400 * t)
)

# Audio Sample 2 (different voice tone)
audio2 = (
    0.5 * np.sin(2 * np.pi * 250 * t) +
    0.3 * np.sin(2 * np.pi * 600 * t)
)

# Add silence
audio1[:3000] = 0
audio2[-3000:] = 0
```

```python
audio1 = audio1 / np.max(np.abs(audio1))
audio2 = audio2 / np.max(np.abs(audio2))
```

```python
audio1_trimmed, _ = librosa.effects.trim(audio1, top_db=20)
audio2_trimmed, _ = librosa.effects.trim(audio2, top_db=20)
```
